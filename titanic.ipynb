```python
enc =LabelEncoder()
titanic['sex'] = enc.fit_transform(titanic['sex'].astype('str'))
titanic['age'] = enc.fit_transform(titanic['age'].astype('str'))
titanic['fare'] = enc.fit_transform(titanic['fare'].astype('str'))
titanic['class'] = enc.fit_transform(titanic['class'].astype('str'))
titanic['embark_town'] = enc.fit_transform(titanic['embark_town'].astype('str'))
titanic['alone'] = enc.fit_transform(titanic['alone'].astype('str'))

titanic.head()
```

```python
mean = X['age'].mean()
mean =round(mean)
X['age'].fillna(mean,inplace = True)
X['age']
```

```python
features = ["age", "fare"]
for feature in features:
    min_ = X[feature].mean() - (3 * X[feature].std())
    max_ = X[feature].mean() + (3 * X[feature].std())
    X = X[X[feature] <= max_]
    X = X[X[feature] >= min_]
    print(feature,    ":", X.shape)
```

```python
features = ["sex", "class", "embark_town", "alone"]
for feature in features:
    count_ = X[feature].value_counts()
    print(feature)
    print(count_, "\n")
```

```python
enc = LabelEncoder()
X["sex"] = enc.fit_transform(X['sex'].astype('str'))
X["class"] = enc.fit_transform(X['class'].astype('str'))
X["embark_town"] = enc.fit_transform(X['embark_town'].\
                                     astype('str'))
X["alone"] = enc.fit_transform(X['alone'].astype('str'))
```
